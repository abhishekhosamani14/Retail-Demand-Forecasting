# 01. Exploratory Data Analysis (EDA)

This notebook performs exploratory analysis on our retail sales dataset. We will analyze missing values, correlations, trends, monthly/weekly seasonality, store-wise and item-wise distributions, and price elasticity.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
os.makedirs("../images", exist_ok=True)

# Load raw data
df = pd.read_csv("../data/train.csv")
df['date'] = pd.to_datetime(df['date'])
print(f"Dataset shape: {df.shape}")
df.head()
```

## 1. Missing Value Analysis

```python
print("Missing values check:")
print(df.isnull().sum())

print("\nData Types:")
print(df.dtypes)
```

## 2. Correlation Analysis

Let's see how numeric columns relate to sales, including our generated `price` column.

```python
plt.figure(figsize=(8, 6))
corr = df[['store', 'item', 'price', 'sales']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".3f", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("../images/correlation_heatmap.png", dpi=150)
plt.show()
```

## 3. Overall Sales Trend Over Time

```python
daily_sales = df.groupby('date')['sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(daily_sales['date'], daily_sales['sales'], color='royalblue', alpha=0.8)
plt.title("Total Daily Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales Units")
plt.tight_layout()
plt.savefig("../images/sales_trend.png", dpi=150)
plt.show()
```

## 4. Monthly Sales Analysis

Let's extract the month and check seasonality.

```python
df['month'] = df['date'].dt.month
monthly_sales = df.groupby('month')['sales'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=monthly_sales, x='month', y='sales', palette='Blues_r')
plt.title("Average Sales by Month (Seasonality check)")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.tight_layout()
plt.savefig("../images/monthly_sales.png", dpi=150)
plt.show()
```

## 5. Store-wise Sales Performance

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='store', y='sales', palette='Set2')
plt.title("Sales Distribution by Store")
plt.xlabel("Store ID")
plt.ylabel("Sales Units")
plt.tight_layout()
plt.savefig("../images/store_sales.png", dpi=150)
plt.show()
```

## 6. Item-wise Sales Performance

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='item', y='sales', palette='Set3')
plt.title("Sales Distribution by Item")
plt.xlabel("Item ID")
plt.ylabel("Sales Units")
plt.tight_layout()
plt.savefig("../images/item_sales.png", dpi=150)
plt.show()
```

## 7. Distribution of Sales and Prices

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['sales'], bins=30, kde=True, ax=axes[0], color='indigo')
axes[0].set_title("Distribution of Sales")
axes[0].set_xlabel("Sales Units")

sns.histplot(df['price'], bins=30, kde=True, ax=axes[1], color='teal')
axes[1].set_title("Distribution of Prices")
axes[1].set_xlabel("Price ($)")

plt.tight_layout()
plt.savefig("../images/distributions.png", dpi=150)
plt.show()
```

## 8. Price Sensitivity / Elasticity Analysis (Demand Curve)

Let's see how sales volume is affected by prices for a specific item (e.g. Item 1).

```python
item_id = 1
item_df = df[df['item'] == item_id]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=item_df, x='price', y='sales', alpha=0.5, color='darkorange')
plt.title(f"Demand Curve for Item {item_id}: Sales vs Price")
plt.xlabel("Price ($)")
plt.ylabel("Sales Units")
plt.tight_layout()
plt.savefig("../images/demand_curve_item1.png", dpi=150)
plt.show()
```
